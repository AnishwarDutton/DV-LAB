```python

import pandas as pd
import matplotlib.pyplot as plt


sentiment_df = pd.read_csv('sentiment_dataset.csv')
print(sentiment_df)

sentiment_df["Text"] = sentiment_df["Text"].fillna("No review provided")


sentiment_df.drop_duplicates(inplace=True)


sentiment_df["Sentiment"] = sentiment_df["Sentiment"].map({"Positive": 1, "Negative": 0})


unique_reviews = sentiment_df["Text"].nunique()
print(f"Number of unique reviews: {unique_reviews}")


sentiment_df["Text"] = sentiment_df["Text"].str.replace("bad", "poor", case=False)


sentiment_df["Review_Length"] = sentiment_df["Text"].str.len()
longest_review = sentiment_df.loc[sentiment_df["Review_Length"].idxmax()]["Text"]
shortest_review = sentiment_df.loc[sentiment_df["Review_Length"].idxmin()]["Text"]
print(f"Longest Review: {longest_review}")
print(f"Shortest Review: {shortest_review}")


review_counts = sentiment_df["Sentiment"].value_counts()
review_counts.plot(kind="bar", color=["blue", "orange"])
plt.title("Number of Positive vs. Negative Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.xticks(ticks=[0, 1], labels=["Negative", "Positive"], rotation=0)
plt.show()

plt.hist(sentiment_df["Review_Length"], bins=30, color="green", alpha=0.7)
plt.title("Distribution of Review Lengths")
plt.xlabel("Review Length")
plt.ylabel("Frequency")
plt.show()
```
