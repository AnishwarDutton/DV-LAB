```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\maheit\Documents\California Wildfire Damage.csv')


df.columns = df.columns.str.strip()


area_burned_by_cause = df.groupby('Cause')['Area_Burned (Acres)'].sum()

plt.figure(figsize=(10, 6))
area_burned_by_cause.plot(kind='bar')
plt.title('Total Area Burned by Cause')
plt.xlabel('Cause')
plt.ylabel('Area Burned (Acres)')
plt.show()
```
